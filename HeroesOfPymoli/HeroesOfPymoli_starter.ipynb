## Most Popular Items

* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Most_Pop_DF = purchase_data.loc[:,["Item ID","Item Name","Price"]]

#Group by item id and item name 
grby_itemID = Most_Pop_DF.groupby(['Item ID','Item Name']) 

#Calculate how many times the items have been purchased
Item_Purchase_Count = grby_itemID['Price'].count()

# How much total revenue generated by each game item
Item_Total_Purchase_Price = grby_itemID['Price'].sum()

# How much is the per item price?
Item_price = Item_Total_Purchase_Price / Item_Purchase_Count

#Display Data frame of Top Popular items. 
Most_Pop_DF = pd.DataFrame({"Purchase Count":Item_Purchase_Count,
                             "Item Price":Item_price,
                             "Total Purchase Value":Item_Total_Purchase_Price})

Sorted_Most_Pop_DF = Most_Pop_DF.sort_values("Purchase Count",ascending=False)

#format using Map function
Sorted_Most_Pop_DF["Item Price"] = Sorted_Most_Pop_DF["Item Price"].map("${:.2f}".format)
Sorted_Most_Pop_DF["Total Purchase Value"] = Sorted_Most_Pop_DF["Total Purchase Value"].map("${:.2f}".format)

Sorted_Most_Pop_DF.head(5)
```
